## Model Building

```python
corpus = []
stemmer = PorterStemmer()
for i in range(0, data.shape[0]):
    reviews = re.sub('[^a-zA-z]', ' ', data.iloc[i]['verified_reviews'])
    reviews = reviews.lower().split()
    reviews = [stemmer.stem(word) for word in reviews if not word in STOPWORDS]
    review = ' '.join(reviews)
    corpus.append(review)
```

```python
cv = CountVectorizer(max_features = 2500)
x = cv.fit_transform(corpus).toarray()
y = data['feedback'].values
```

```python
pickle.dump(cv, open('models/cv.pkl', 'wb'))
```
